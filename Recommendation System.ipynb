```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
movie_titles_df = pd.read_csv('Movie_Id_Titles')
```

```python
movie_titles_df
```

```python
rating_df = pd.read_csv('u.data', sep ='\t', names = ['user_id', 'item_id', 'rating', 'timestamp'])
```

```python
rating_df.drop(['timestamp'], axis=1, inplace = True)
```

```python
rating_df
```

```python
rating_df.info()
```

```python
movies_rating_df = pd.merge(movie_titles_df, rating_df, on = 'item_id')
```

```python
movies_rating_df
```

```python
movies_rating_df.groupby('title').describe()
```

```python
movies_rating_df.groupby('title')['rating'].describe()
```

```python
df_rating_mean = movies_rating_df.groupby('title')['rating'].describe()['mean']
```

```python
df_rating_counts = movies_rating_df.groupby('title')['rating'].describe()['count']
```

```python
df_rating_counts
```

```python
df = pd.concat([df_rating_mean, df_rating_counts], axis = 1)
```

```python
df
```

```python
df = df.reset_index()
```

```python
df
```

```python
df['mean'].plot(bins = 100, kind = 'hist', color = 'b' )
```

```python
df['count'].plot(bins = 100, kind = 'hist', color = 'r' )
```

```python
titanic_corr = pd.DataFrame(ratings_user_matrix.corrwith (titanic), columns = ['Correlation'])
```
